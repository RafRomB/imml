```python
from utils import RandomMultiviewDataset
from decomposition import MOFA
from algorithms import MOFAClustering
```

```python
rmd = RandomMultiviewDataset()
mvd = rmd.create_mvd(n_samples = 100, n_features = [10, 50, 100], n_clusters = [3, 4, 2])
sample_views = rmd.add_sample_views(mvd = mvd, p = [0.2, 0.3, 0.5])
imvd = rmd.add_missing_views(mvd = mvd, sample_views = sample_views)
for data in imvd:
    print(data.shape)
    display(data.head())
```

```python
from sklearn.pipeline import make_pipeline
from utils import FillMissingViews
```

```python
mofa_pipeline = make_pipeline(FillMissingViews(value="nan"), MOFA().set_output(transform = 'pandas'))
```

```python
mofa_pipeline.fit(X = imvd)
```

```python
transformed_imvd = mofa_pipeline.transform(X = imvd)
print(transformed_imvd.shape)
display(transformed_imvd.head())
```
