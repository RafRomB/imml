```python
import numpy as np
import pandas as pd
from utils.utils import RandomMultiviewDataset, FillMissingViews, get_sample_views, ConcatenateViews, DropView, SingleView
from sklearn.pipeline import make_pipeline
from IPython.display import display
```

```python
rmd = RandomMultiviewDataset()
mvd = rmd.create_mvd(n_samples = 100, n_features = [10, 50, 100], n_clusters = [3, 4, 2])
for data in mvd:
    display(data.head())
```

```python
sample_views = rmd.add_sample_views(mvd = mvd, p = [0.2, 0.3, 0.5])
sample_views
```

```python
imvd = rmd.add_missing_views(mvd = mvd, sample_views = sample_views)
for data in imvd:
    display(data.head())
```

```python
get_sample_views(imvd= imvd)
```

```python
filled_imvd = FillMissingViews(value= 'zeros').transform(X = imvd)
for data in filled_imvd:
    display(data.head())
```

```python
filled_imvd = FillMissingViews(value= 'mean').fit_transform(X = imvd)
for data in filled_imvd:
    display(data.head())
```

```python
drop_view_imvd = DropView(view_idx= 1).transform(X = imvd)
for data in drop_view_imvd:
    display(data.head())
```

```python
sv = SingleView(view_idx= 1).transform(X = imvd)
sv.head()
```

```python
new_mvd = FillMissingViews(value= 'mean').fit_transform(X = imvd)
ConcatenateViews().transform(X = new_mvd)
```

```python
make_pipeline(FillMissingViews(value= 'mean'), ConcatenateViews()).fit_transform(X = imvd)
```
