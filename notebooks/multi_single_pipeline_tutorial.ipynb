```python
from utils.utils import RandomMultiviewDataset
from utils import MultiSinglePipeline
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
```

We create a random multi-view dataset with 10 samples.

```python
rmd = RandomMultiviewDataset()
mvd = rmd.create_mvd(n_samples = 100, n_features = [10, 50, 100], n_clusters = [3, 4, 2])
sample_views = rmd.add_sample_views(mvd = mvd, p = [0.2, 0.3, 0.5])
imvd = rmd.add_missing_views(mvd = mvd, sample_views = sample_views)
for data in imvd:
    display(data.head())
```

```python
estimator = MultiSinglePipeline(pipeline = make_pipeline(SimpleImputer().set_output(transform = 'pandas'), StandardScaler().set_output(transform = 'pandas'), KMeans()))
estimator.fit(X = imvd)
```

Now we will remove some views from the samples.

```python
estimator.pipelines_
```

```python
estimator.predict(X = imvd)
```
