```python
from imvc.pipelines import MSVPipeline
from imvc.datasets import LoadDataset
from imvc.utils import DatasetUtils
```

We create a random multi-view dataset with 10 samples.

```python
Xs = LoadDataset.load_incomplete_nutrimouse(p = 0.2)
for X in Xs:
    print(X.shape)
    display(X.head())
```

```python
samples = DatasetUtils.get_sample_names(Xs=Xs)
estimator = MSVPipeline(n_clusters = 3, samples= samples)
estimator.fit(Xs)
```

```python
estimator.pipeline_list_[0]
```

```python
estimator.predict(Xs)
```
