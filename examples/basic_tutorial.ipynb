```python
from transformers import FillMissingViews, DropView, SingleView, ConvertToNM, MultiViewTransformer
from sklearn.preprocessing import StandardScaler, RobustScaler
from datasets import load_incomplete_nutrimouse
from sklearn.pipeline import make_pipeline
from IPython.display import display
from mvlearn.datasets import load_nutrimouse
```

```python
Xs = load_incomplete_nutrimouse(p = 0.2)
for X in Xs:
    print(X.shape)
    display(X.head())
```

```python
transformer = MultiViewTransformer(transformer = ConvertToNM())
transformed_Xs = transformer.fit_transform(Xs)
for X in transformed_Xs:
    print(X.shape)
    display(X.head())
```

```python
FillMissingViews(value= 'mean').fit(Xs).features_view_mean_list_
```

```python
drop_view_imvd = DropView(view_idx= 1).transform(X = Xs)
for data in drop_view_imvd:
    display(data.head())
```

```python
sv = SingleView(view_idx= 1).transform(X = Xs)
sv.head()
```

```python
make_pipeline(FillMissingViews(value= 'mean'), SingleView(view_idx= 1)).fit_transform(Xs).head()
```

```python
pipeline = make_pipeline(
    FillMissingViews(value= 'mean'),
    MultiViewTransformer(transformer = [StandardScaler().set_output(transform = 'pandas'),
                                        RobustScaler().set_output(transform = 'pandas')]),
    SingleView(view_idx= 1))
pipeline.fit_transform(Xs).head()
```
