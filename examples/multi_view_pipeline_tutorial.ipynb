```python
from imvc.pipelines import MultiViewPipeline
from sklearn.preprocessing import StandardScaler
from imvc.datasets import LoadDataset
from IPython.display import display
from mvlearn.datasets import load_nutrimouse
from sklearn.cluster import KMeans
```

We create a random multi-view dataset with 10 samples.

```python
Xs = LoadDataset.load_incomplete_nutrimouse(p = 0.2)
for X in Xs:
    print(X.shape)
    display(X.head())
```

```python
estimator = MultiViewPipeline(steps = [StandardScaler().set_output(transform = 'pandas'), KMeans()])
estimator.fit(Xs)
```

Now we will remove some views from the samples.

```python
estimator.pipeline_list_
```

```python
estimator.predict(Xs)
```
